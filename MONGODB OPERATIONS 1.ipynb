### AGGREGATION PIPELINES
1. Create an aggregation pipeline that will display the total number of employees
for each role in the empl collection

```python
EMP> db.empl.aggregate([ { $group: { _id: "$role", totalEmployees: { $sum: 1 } } }] )
[
  { _id: 'ALR', totalEmployees: 2 },
  { _id: 'BW', totalEmployees: 2 },
  { _id: 'OB', totalEmployees: 3 },
  { _id: 'C and WK', totalEmployees: 1 },
  { _id: 'MOB', totalEmployees: 2 }
]
```

2. Create an aggregation pipeline that will display the total salary for each role in
the empl collection, sorted by total salary in ascending order.

```python

EMP> db.empl.aggregate([
... {
... $group: {
... _id: "$role",
... totalSalary: { $sum: "$salary" }
... }
... },
... {
... $sort: { totalSalary: 1 }
... }
... ])
[
  { _id: 'C and WK', totalSalary: 9500 },
  { _id: 'BW', totalSalary: 16700 },
  { _id: 'ALR', totalSalary: 17400 },
  { _id: 'MOB', totalSalary: 17500 },
  { _id: 'OB', totalSalary: 32800 }
]
```
